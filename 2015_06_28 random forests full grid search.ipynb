```python
X_df = pd.read_pickle('new_X_df.save')
outcomes_df = pd.read_pickle('new_outcomes_df.save')
```

```python
n_samples = np.shape(X_df)[0]
shuffle_i = np.random.permutation(n_samples)
```

```python
X = X_df.ix[shuffle_i,:]
y = outcomes_df.ix[shuffle_i,'Outcome'].values
```

```python
# rf = RandomForestClassifier()

# parameters = {'n_estimators':[800,900,1000],
#               'max_depth':[20,22,24,None],
#               'min_samples_split':[2,3,4],
#               'min_samples_leaf':[1,2]}
```

```python
rf = RandomForestClassifier()

parameters = {'n_estimators':[800,900,1000],
              'max_depth':[22,24,None],
              'min_samples_split':[2,3,4],
              'min_samples_leaf':[1,2]}
```

```python
clf_grid = grid_search.GridSearchCV(rf,param_grid=parameters,\
                            scoring='roc_auc',n_jobs=-1, verbose=1)
```

```python
clf_grid.fit(X, y)
```

```python
for score in clf_grid.grid_scores_:
    print score
```

```python
clf_grid.best_params_
```

```python
clf_grid.best_score_
```

```python
rf_best = clf_grid.best_estimator_
```
